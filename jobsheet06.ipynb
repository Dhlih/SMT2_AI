## 1. Melakukan Visualisasi Data menggunakan Pie Chart

```python
import matplotlib.pyplot as plt

flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (12, 11, 4, 8, 7)
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D','#D53032')
explode = (0, 0, 0, 0.1, 0)

plt.pie(votes, labels=flavors, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True)
plt.show()

```

## 2. Melakukan Visualisasi Data Menggunakan Bar Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

countries = ('Argentina', 'Bolivia', 'Brazil', 'Chile',  'Colombia', 'Ecuador', 'Falkland Islands', 'French Guiana', 'Guyana', 'Paraguay', 'Peru','Suriname', 'Uruguay', 'Venezuela')
populations = (45076704, 11626410, 212162757, 19109629,50819826, 17579085, 3481, 287750, 785409, 7107305, 32880332,585169, 3470475, 28258770)

df = pd.DataFrame({
    'Country' : countries,
    'Population' : populations,
})

df = df.sort_values(by='Population')
# menentukan posisi bar di mana saja dengan mengambil index
x_coords = np.arange(len(countries))
# membuat bar di index menggunakan data population
plt.bar(x_coords, df["Population"], tick_label=df["Country"])
plt.xticks(rotation=90)
plt.ylabel('Population')
plt.title('South American Populations')

print(plt.show())


```

```python
import matplotlib.pyplot as plt

candy_names = ['Kit Kat', 'Snickers', 'Milky Way', 'Toblerone','Twix']
candy_counts = [52, 39, 78, 13, 78]

total = sum(candy_counts)  # 52 + 39 + 78 + 13 + 78 = 260

snickers_prob = candy_counts[1] / total  # 39/260 = 0.15
print(f"Peluang mendapatkan Snickers: {snickers_prob:.2f} atau {snickers_prob*100:.1f}%")

colors = ['#F4B400', '#4285F4', '#DB4437', '#0F9D58', '#FF6D00']

explode = (0, 0.1, 0, 0, 0)

plt.figure(figsize=(8, 6))
plt.pie(candy_counts,
        labels=candy_names,
        autopct='%1.1f%%',
        colors=colors,
        explode=explode,
        shadow=True)
plt.title('Persentase Permen dalam Kantong Mario')
plt.show()
```

## Latihan 3:Makanan

```python
import matplotlib.pyplot as plt

# Data
dessert_sales = {
    'Lava Cake': 14, 'Mousse': 5, 'Chocolate Cake': 12,
    'Ice Cream': 19, 'Truffles': 6, 'Brownie': 8,
    'Chocolate Chip Cookie': 12, 'Chocolate Pudding': 9,
    'Souffle': 10, 'Chocolate Cheesecake': 17,
    'Chocolate Chips': 2, 'Fudge': 9, 'Mochi': 13
}

# Urutkan dari yang paling tidak populer
sorted_sales = sorted(dessert_sales.items(), key=lambda x: x[1])
desserts = [item[0] for item in sorted_sales]
sales = [item[1] for item in sorted_sales]

# Buat bar chart horizontal
plt.figure(figsize=(10, 6))
bars = plt.barh(desserts, sales, color='purple')

# Highlight 3 terendah
for i in range(3):
    bars[i].set_color('yellow')

plt.title('Popularitas Makanan Penutup (Penjualan Mingguan)')
plt.xlabel('Jumlah Terjual')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
